```python
import tablemage as tm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets

# load the iris dataset
iris = sklearn.datasets.load_iris(as_frame=True)
iris = iris.frame
iris
```

```python
sns.set_theme(context="poster", style="whitegrid")

# plot the distribution of the iris dataset by species as a single histogram, only with the kdensity estimate
fig, ax = plt.subplots()

# set line width
sns.kdeplot(
    data=iris, x="sepal length (cm)", hue="target", ax=ax, legend=False, linewidth=7
)

# get rid of gridlines
ax.grid(False)
ax.set_xlabel("")
ax.set_ylabel("")

# get rid of gray box
ax.set_facecolor("white")
# get rid of all tick marks
ax.tick_params(
    axis="both",  # changes apply to the x-axis
    which="both",  # both major and minor ticks are affected
    bottom=False,  # ticks along the bottom edge are off
    top=False,
    left=False,
    right=False,
    labelbottom=False,
    labelleft=False,
)
# get rid of gray bounding box
sns.despine(left=True, bottom=True)
# fig.tight_layout()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Generate data
x = np.linspace(-4, 5, 1000)
density_A = np.exp(-0.5 * ((x + 1) / 1) ** 2) / (np.sqrt(2 * np.pi) * 1)
density_B = np.exp(-0.5 * ((x - 1) / 1) ** 2) / (np.sqrt(2 * np.pi) * 1)

fig, ax = plt.subplots()
ax.plot(x, density_A, color="purple", label="Distribution A", linewidth=7)
ax.plot(x, density_B, color="black", label="Distribution B", linewidth=7)
ax.set_xlim(-4, 4)
# get rid of gray box
ax.set_facecolor("white")
# get rid of all tick marks
ax.tick_params(
    axis="both",  # changes apply to the x-axis
    which="both",  # both major and minor ticks are affected
    bottom=False,  # ticks along the bottom edge are off
    top=False,
    left=False,
    right=False,
    labelbottom=False,
    labelleft=False,
)
sns.despine(left=True, bottom=True)
ax.grid(False)
# Show plot
plt.show()
```

```python
# plot the distribution of iris dataset, simple scatterplot

fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(
    data=iris,
    x="sepal length (cm)",
    y="sepal width (cm)",
    hue="target",
    ax=ax,
    s=100,
    legend=False,
    edgecolor="black",
)
ax.set_xlabel("")
ax.set_ylabel("")
# get rid of gray box
ax.set_facecolor("white")
# get rid of all tick marks
ax.tick_params(
    axis="both",  # changes apply to the x-axis
    which="both",  # both major and minor ticks are affected
    bottom=False,  # ticks along the bottom edge are off
    top=False,
    left=False,
    right=False,
    labelbottom=False,
    labelleft=False,
)
sns.despine(left=True, bottom=True)
ax.grid(False)
# Show plot
plt.show()
```
